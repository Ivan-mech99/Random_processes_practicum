# Задача 3

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook
```

```python
N = 10001
```

```python
def predict():
    hold = np.random.rand()
    if hold < 0.5:
        return 1
    else:
        return -1
    
def random_walk(N):
    res1 = 0
    data = np.zeros(N, dtype='int64')
    for i in range(1, N):
        hold = predict()
        data[i] = data[i-1] + hold
    for i in range(0, N-1):
        if data[i] > 0 or data[i+1] > 0:
            res1 = res1 + 1
    return res1 / (N-1)
```

```python
N1 =10000
results = []
for i in tqdm_notebook(range(0, N1)):
    hold = random_walk(N)
    results.append(hold)
```

```python
data1 = np.array(results)
resik1 = np.array((data1)*(N-1), dtype = 'int64')
c = np.bincount(resik1)
```

```python
plt.figure(figsize=(20,14))
plt.title("график времени на положительной полуоси", fontsize=20)
plt.xlabel('Доля времени', fontsize=20)
plt.ylabel('Число процессов', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(np.arange(0,N)/(N-1),c)
plt.show()
```

# Задача 4

```python
N = 10001
```

```python
def random_walk_simple(N):
    res1 = 0
    data = np.zeros(N, dtype='int64')
    for i in range(1, N):
        hold = predict()
        data[i] = data[i-1] + hold
    z = np.where(data == 0)[0]
    res1 = z[z.shape[0]-1]
    return res1
```

```python
N1 =10000
results = []
for i in tqdm_notebook(range(0, N1)):
    hold = random_walk_simple(N)
    results.append(hold)
```

```python
c1 = np.bincount(np.array(results))
```

```python
plt.figure(figsize=(20,14))
plt.title("Распределение момента последнего возвращения в 0", fontsize=20)
plt.xlabel('Момент последнего возвращения в 0', fontsize=20)
plt.ylabel('Число процессов', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(np.arange(1,N),c1[1::])
plt.show()
```
