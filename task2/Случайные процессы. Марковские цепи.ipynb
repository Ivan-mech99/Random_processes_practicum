# Задача 3

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat=np.array([[0.6,0.2,0.2],[0.05,0.9,0.05],[0.1,0.2,0.7]])
```

```python
def predict(arr):
    hold = np.random.rand()
    if hold < arr[0]:
        return 0
    if (hold >= arr[0]) and (hold < arr[0]+arr[1]):
        return 1
    if hold >= arr[0]+arr[1] and (hold < 1):
        return 2
```

```python
x1 = np.random.rand()
x2 = np.random.rand()*(1.0-x1)
x3 = 1.0 - x1 -x2
init = np.array([x1, x2, x3])
```

```python
init
```

```python
results = np.zeros((1001,),dtype='int64')
results[0] = predict(init)
for i in range(1,1001):
    results[i] = predict(mat[results[i-1]])
```

```python
plt.figure(figsize=(14,10))
plt.plot(np.arange(0,1001),results, color='green', marker='o', markersize=3)
plt.xlabel('Number of step', fontsize=20)
plt.ylabel('Position', fontsize=20)
plt.grid()
plt.show()
```
